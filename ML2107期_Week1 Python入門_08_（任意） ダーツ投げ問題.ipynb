【問題5】関数化

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches


# ダーツのまとの数
target_num = 6

# ダーツを投げる回数
throw_num = 1000


def make_darts_place(target_num, throw_num):

#    darts_place = {}
    array_throw_dice = []
    array_darts_coordinate_x = []
    array_darts_coordinate_y = []

    for i in range(throw_num - 1):
        throw_dice = np.random.choice(range(1, target_num + 1, 1)) # ダーツのまとの番号
        darts_coordinate_x = np.random.normal(throw_dice, 0.1)
        darts_coordinate_y = np.random.normal(0, 0.1)

        array_throw_dice.append(throw_dice)
        array_darts_coordinate_x.append(darts_coordinate_x)
        array_darts_coordinate_y.append(darts_coordinate_y)

    return array_throw_dice, array_darts_coordinate_x, array_darts_coordinate_y

array_throw_dice, array_darts_coordinate_x, array_darts_coordinate_y = make_darts_place(target_num, throw_num)

figure, axes = plt.subplots(figsize=(20.0,15.0))
ax = plt.axes()

for i in range(target_num):
    ax.add_patch(patches.Circle(xy=(i+1, 0), radius=0.5, fc="w", ec="b"))
    ax.add_patch(patches.Circle(xy=(i+1, 0), radius=0.3, fc="w", ec="b"))
    ax.add_patch(patches.Circle(xy=(i+1, 0), radius=0.1, fc="w", ec="b"))

for i in range(len(array_throw_dice)):
    ax.plot(array_darts_coordinate_x[i], array_darts_coordinate_y[i], marker=".")

plt.axis("scaled")
#plt.axis("equal")

ax.set_xlim(0, target_num + 1)
ax.set_ylim(-0.7, 0.7)

#ax.set_aspect(1.0)

plt.show()
```

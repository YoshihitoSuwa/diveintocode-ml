【問題6】図の作成<br>
　① ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）<br>
　② 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する<br>
　③ 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する<br>

```python
import matplotlib.pyplot as plt


print("")
print("① ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）")
print("")
# ラベル：sepal_length
df_value_cnt = df["sepal_length"].value_counts()
plt.title("sepal_length")
plt.pie(df_value_cnt, autopct="%.1f%%")
plt.show()
# ラベル：sepal_width
df_value_cnt = df["sepal_width"].value_counts()
plt.title("sepal_width")
plt.pie(df_value_cnt, autopct="%.1f%%")
plt.show()
# ラベル：petal_length
df_value_cnt = df["petal_length"].value_counts()
plt.title("petal_length")
plt.pie(df_value_cnt, autopct="%.1f%%")
plt.show()
# ラベル：petal_width
df_value_cnt = df["petal_width"].value_counts()
#label = [str(i) for i in df_value_cnt.index]
plt.title("petal_width")
#plt.pie(df_value_cnt, labels=label, autopct="%.1f%%")
plt.pie(df_value_cnt, autopct="%.1f%%")
#plt.legend(loc="upper right")
plt.show()
# ラベル：Species
df_value_cnt = df["Species"].value_counts()
#label = [str(i) for i in df_value_cnt.index]
plt.title("Species")
#plt.pie(df_value_cnt, labels=label, autopct="%.1f%%")
plt.pie(df_value_cnt, autopct="%.1f%%")
#plt.legend()
plt.show()
```

```python
print("")
print("② 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する")
print("")
sepal_length_1 = df[df["Species"]==0].loc[:,"sepal_length"]
sepal_length_2 = df[df["Species"]==1].loc[:,"sepal_length"]
sepal_length_3 = df[df["Species"]==2].loc[:,"sepal_length"]
sepal_width_1 = df[df["Species"]==0].loc[:,"sepal_width"]
sepal_width_2 = df[df["Species"]==1].loc[:,"sepal_width"]
sepal_width_3 = df[df["Species"]==2].loc[:,"sepal_width"]
petal_length_1 = df[df["Species"]==0].loc[:,"petal_length"]
petal_length_2 = df[df["Species"]==1].loc[:,"petal_length"]
petal_length_3 = df[df["Species"]==2].loc[:,"petal_length"]
petal_width_1 = df[df["Species"]==0].loc[:,"petal_width"]
petal_width_2 = df[df["Species"]==1].loc[:,"petal_width"]
petal_width_3 = df[df["Species"]==2].loc[:,"petal_length"]

# ラベル：sepal_length、x軸：Species、y軸：sepal_length
plt.title("sepal_length")
plt.xlabel("Species")
plt.ylabel("sepal_length")
plt.boxplot([sepal_length_1, sepal_length_2, sepal_length_3])
plt.show()

# ラベル：sepal_width、x軸：Species、y軸：sepal_width
plt.title("sepal_width")
plt.xlabel("Species")
plt.ylabel("sepal_width")
plt.boxplot([sepal_width_1, sepal_width_2, sepal_width_3])
plt.show()

# ラベル：petal_length、x軸：Species、y軸：petal_length
plt.title("petal_length")
plt.xlabel("Species")
plt.ylabel("petal_length")
plt.boxplot([petal_length_1, petal_length_2, petal_length_3])
plt.show()

# ラベル：petal_width、x軸：Species、y軸：petal_length
plt.title("petal_width")
plt.xlabel("Species")
plt.ylabel("petal_length")
plt.boxplot([petal_width_1, petal_width_2, petal_width_3])
plt.show()
```

```python
print("")
print("③ 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する")
print("")

# ラベル：sepal_length、x軸：Species、y軸：sepal_length
plt.title("sepal_length")
plt.xlabel("Species")
plt.ylabel("sepal_length")
plt.violinplot([sepal_length_1, sepal_length_2, sepal_length_3])
plt.show()

# ラベル：sepal_width、x軸：Species、y軸：sepal_width
plt.title("sepal_width")
plt.xlabel("Species")
plt.ylabel("sepal_width")
plt.violinplot([sepal_width_1, sepal_width_2, sepal_width_3])
plt.show()

# ラベル：petal_length、x軸：Species、y軸：petal_length
plt.title("petal_length")
plt.xlabel("Species")
plt.ylabel("petal_length")
plt.violinplot([petal_length_1, petal_length_2, petal_length_3])
plt.show()

# ラベル：petal_width、x軸：Species、y軸：petal_length
plt.title("petal_width")
plt.xlabel("Species")
plt.ylabel("petal_length")
plt.violinplot([petal_width_1, petal_width_2, petal_width_3])
plt.show()
```

【問題1】乱数の作成

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)

mean = (-3, 0)
cov = [[1.0, 0.8],[0.8, 1.0]]

x,y = np.random.multivariate_normal(mean, cov, 500).T

plt.plot(x, y, "x")
plt.show()

#x,y = np.random.multivariate_normal(mean, cov, 500).T
#plt.plot(x, y, "x")
#plt.axis("equal")
#plt.show()
```

【問題3】ヒストグラムによる可視化

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)

mean = (-3, 0)
cov = [[1.0, 0.8],[0.8, 1.0]]

x,y = np.random.multivariate_normal(mean, cov, 500).T

plt.xlim(-6,3)
plt.hist(x, color = "blue")
plt.hist(y, color = "orange")
plt.show()
```

【問題4】データの追加

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)

mean_0 = (-3, 0)
cov_0 = [[1.0, 0.8],[0.8, 1.0]]

x_0,y_0 = np.random.multivariate_normal(mean_0, cov_0, 500).T

mean_1 = (0, -3)
cov_1 = [[1.0, 0.8],[0.8, 1.0]]

x_1,y_1 = np.random.multivariate_normal(mean_1, cov_1, 500).T

plt.scatter(x_0, y_0, label="0")
plt.scatter(x_1, y_1, label="1")
plt.legend()
plt.show()
```

【問題5】データの結合

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)

mean_0 = (-3, 0)
cov_0 = [[1.0, 0.8],[0.8, 1.0]]

x_0,y_0 = np.random.multivariate_normal(mean_0, cov_0, 500).T

mean_1 = (0, -3)
cov_1 = [[1.0, 0.8],[0.8, 1.0]]

x_1,y_1 = np.random.multivariate_normal(mean_1, cov_1, 500).T

xy_0 = np.concatenate([x_0, y_0])
xy_1 = np.concatenate([x_1, y_1])
xy01 = np.stack([xy_0, xy_1], 1)
print(xy01.shape)
```

```python
zero_0 = np.zeros(500)
one_1 = np.ones(500)
label = np.concatenate([zero_0, one_1])
xy01_label = np.stack([xy_0, xy_1, label], 1)
print(xy01_label.shape)
```

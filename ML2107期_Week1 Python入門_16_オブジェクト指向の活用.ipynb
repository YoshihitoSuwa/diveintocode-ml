## 【問題4】 四則演算を行うクラスの作成
```
class ExampleClass():
    """
    説明用の簡単なクラス

    ----------
    value : float or int
        初期値
    Attributes
    ----------

    """
    def __init__(self, value):
        self.value = value
        print("初期値{}が設定されました".format(self.value))

        """
        self.value += value2
example = ExampleClass(5)
print("value : {}".format(example.value))

```

上記ExampleClassは足し算のメソッドを持っていますが、これに引き算、掛け算、割り算のメソッドを加えてください。<br>
コンストラクタに入力されたvalueが文字列や配列など数値以外だった場合にはエラーを出すようにしてください。<br>
クラス名や説明文も適切に書き換えてください。<br>

````python
class ExampleClass():
    """
    四則演算のためのクラス
    Parameters
    ----------
    value : float or int
        初期値
    Attributes
    ----------
    value : float or int
        計算結果
    """
    def __init__(self, value):
        self.value = value
        print("初期値{}が設定されました".format(self.value))
    def add(self, value2):
        """
        受け取った引数をself.valueに加える
        """
        # int型以外は「Error!」と表示
        if type(value2) is int:
            self.value += value2
        else:
            print("Error!")
    def sub(self, value2):
        """
        受け取った引数をself.valueから引く
        """
        # int型以外は「Error!」と表示
        if type(value2) is int:
            self.value -= value2
        else:
            print("Error!")
    def mul(self, value2):
        """
        受け取った引数をself.valueに掛ける
        """
        # int型以外は「Error!」と表示
        if type(value2) is int:
            self.value *= value2
        else:
            print("Error!")
    def div(self, value2):
        """
        受け取った引数をself.valueから割る
        """
        # int型以外は「Error!」と表示
        if type(value2) is int:
            self.value /= value2
        else:
            print("Error!")

        
example = ExampleClass(5)
print("value : {}".format(example.value))
example.add(3)
print("value : {}".format(example.value))
example.sub(3)
print("value : {}".format(example.value))
example.mul(3)
print("value : {}".format(example.value))
example.div(3)
print("value : {}".format(example.value))
````

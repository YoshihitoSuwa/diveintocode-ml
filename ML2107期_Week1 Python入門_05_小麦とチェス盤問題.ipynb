【問題1】2×2マスのチェス盤の小麦の数

```python
import numpy as np


n_squares = 4
small_board_list = [1]
for _ in range(n_squares - 1):
    small_board_list.append(2*small_board_list[-1])
print("4マスの板に小麦を並べる（リスト）：{}".format(small_board_list))

small_board_ndarray = np.array(small_board_list)
small_board_ndarray = small_board_ndarray.reshape(2, 2)
print("4マスの板に小麦を並べる（ndarray）：{}".format(small_board_ndarray))
```

【問題2】n×mマスへの拡張

```python
import numpy as np


# 入力値チェック用関数
def check_input(item):
    check_int = 0
    while check_int == 0:
        num = input("{}を入力してください：".format(item))
        if num.isdecimal():
            num = int(num)
            check_int = 1
        else:
            print("整数を入力してください。")
    return num


# n×mマスへのreshape用関数
def make_ndarray(board_list, n_line, m_column):
    board_list = np.array(board_list)
    board_list = board_list.reshape(n_line, m_column)
    return board_list


# 行数(n)と列数(m)の入力
n_line = check_input("行数")
m_column = check_input("列数")

# 入力された値を基に、各マスへ置く小麦の数の配列を作成
n_squares = n_line * m_column
board_list = [1]
for _ in range(n_squares - 1):
    board_list.append(2*board_list[-1])

board_list = make_ndarray(board_list, n_line, m_column)
print(board_list)
```

【問題3】小麦の数の合計

```python
import numpy as np
import matplotlib.pyplot as plt


# n×mマスへのreshape用関数
def make_ndarray(board_list, n_line, m_column):
    board_list = np.array(board_list)
    board_list = board_list.reshape(n_line, m_column)
    return board_list


# 行数(8)と列数(8)の定義
N_LINE = 8
M_COLUMN = 8

# 入力された値を基に、各マスへ置く小麦の数の配列を作成
n_squares = N_LINE * M_COLUMN
board_list = [1]
for _ in range(n_squares - 1):
    board_list.append(2*board_list[-1])

board_list = make_ndarray(board_list, N_LINE, M_COLUMN)
print("8x8マスのチェス盤の小麦の合計：{}個".format(board_list.sum()))

%matplotlib inline
plt.xlabel("column")
plt.ylabel("number")
plt.title("number in each column")
plt.bar(np.arange(1,9), board_list.mean(axis=1))
plt.show()
```
